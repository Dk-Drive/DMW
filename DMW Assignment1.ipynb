```python
df = pd.read_csv("USA_cars_datasets.csv")
```

```python
# Data Reduction
subset = df[['price','brand','model','year','mileage','color','state','country']]
subset.head()
```

```python
# Data  Cleaning
subset.isnull().sum()
```

```python
subset2 = df[['title_status','model','lot','vin']]
subset2.head()
```

```python
# Data Integration
subset3 = pd.merge(subset,subset2,on='model')
subset3.head()
```

```python
# Data Transformation
subset3.to_excel('Car_data.xlsx')
```
